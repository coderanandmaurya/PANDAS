```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/coderanandmaurya/PANDAS/main/dropduplicatesexample.csv')
df
```

```python
#check whole row for duplicacy # check 5th line 
new_df=df.drop_duplicates()
new_df
```

```python
#drop rows which match on columns A and C
sub_df= df.drop_duplicates(subset=['A','C'])
sub_df  
```
