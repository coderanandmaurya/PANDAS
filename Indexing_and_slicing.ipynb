```python
import pandas as pd
```

```python
# read from raw data
URL="https://raw.githubusercontent.com/coderanandmaurya/PANDAS/main/matches.csv"
df=pd.read_csv(URL)
df.head(4)
```

## **To access single column by column name**

```python
df["city"]
```

## **To access Multiple column by column name**

```python
df[["city","winner"]]
```


## **Slicing columns in data frame**

```python
df.iloc[:,0:2]
```

## **Slicing rows in data frame**

```python
df.iloc[0:3]
```

```python
df.iloc[0:3,0:6]
```

## **Fancy indexing**

```python
df.iloc[:,[1,2,3]]
```
